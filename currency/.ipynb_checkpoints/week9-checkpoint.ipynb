## Time Series
A time series is a series of data points indexed (or listed or graphed) in time order. Most commonly, a time series is a sequence taken at successive equally spaced points in time. Thus it is a sequence of discrete-time data. 
- Trend 
- Seasonality
- cyclicity
- Irregular
---- 
##### Tools: Statsmodels
from statsmodels.api import * 
#### 1. AutoRegressive (AR)
$ X_t = \phi_1 x_{t-1} + \phi_2 x_{t-2} + \phi_3 x_{t-3} +  ...  + \phi_p x_{t-p} + u_t $
- u_t is the pure noise


#### 2. MovingAverage(MA) 
$ X_t = u_t + \phi_1 x_{t-1}+ \phi_2 x_{t-2} + .. + \phi_q x_{t-q} $ 


#### 3. ARMA 
$ ARMA(p,q):   X_t = u_t + \phi_1 x_{t-1}+ \phi_2 x_{t-2} + ...  + \phi_p x_{t-p} + \sigma_1 x_t-1 + \sigma_2 x_t-2 + \sigma_1 x_t-1 $ 

#### 4. ARIMA (intergrated)



-----
##### Evalutaion methods: AIC rules / BIC

#### ARMA stock predicting:

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_model import ARMA
import warnings
from itertools import product
from datetime import datetime, timedelta
import calendar
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('./currency_data/shanghai_index_1990_12_19_to_2020_03_12.csv')
df = df[['Timestamp', 'Price']]


df.Timestamp = pd.to_datetime(df.Timestamp)
df.index = df.Timestamp
df
```

```python
df.describe
```

```python
df_month = df.resample('M').mean()
df_Q = df.resample('Q-DEC').mean()
df_year = df.resample('A-DEC').mean()
fig = plt.figure(figsize=[15, 7])
plt.rcParams['font.sans-serif']=['SimHei'] #ç”¨æ¥æ­£å¸¸æ˜¾ç¤ºä¸­æ–‡æ ‡ç­¾
plt.suptitle('æ²ªå¸‚æŒ‡æ•°', fontsize=20)
plt.subplot(221)
plt.plot(df.Price, '-', label='æŒ‰å¤©')
plt.legend()
plt.subplot(222)
plt.plot(df_month.Price, '-', label='æŒ‰æœˆ')
plt.legend()
plt.subplot(223)
plt.plot(df_Q.Price, '-', label='æŒ‰å­£åº¦')
plt.legend()
plt.subplot(224)
plt.plot(df_year.Price, '-', label='æŒ‰å¹´')
plt.legend()
plt.show()
```

```python
ps = range(0, 3)
qs = range(0, 3)
parameters = product(ps, qs)
parameters_list = list(parameters)
print(parameters_list)
# å¯»æ‰¾æœ€ä¼˜ARMAæ¨¡åž‹å‚æ•°ï¼Œå³best_aicæœ€å°
results = []
best_aic = float("inf") # æ­£æ— ç©·
for param in parameters_list:
    try:
        model = ARMA(df_month.Price,order=(param[0], param[1])).fit()
    except ValueError:
        print('å‚æ•°é”™è¯¯:', param)
        continue
    aic = model.aic
    if aic < best_aic:
        best_model = model
        best_aic = aic
        best_param = param
    results.append([param, model.aic])
# è¾“å‡ºæœ€ä¼˜æ¨¡åž‹
print('æœ€ä¼˜æ¨¡åž‹: ', best_model.summary())
```

```python
df_month2 = df_month[['Price']]
future_month = 3
last_month = pd.to_datetime(df_month2.index[len(df_month2)-1])
#print(last_month)
date_list = []
for i in range(future_month):
    # è®¡ç®—ä¸‹ä¸ªæœˆæœ‰å¤šå°‘å¤©
    year = last_month.year
    month = last_month.month
    if month == 12:
        month = 1
        year = year+1
    else:
        month = month + 1
    next_month_days = calendar.monthrange(year, month)[1]
    #print(next_month_days)
    last_month = last_month + timedelta(days=next_month_days)
    date_list.append(last_month)
print('date_list=', date_list)
```

```python

# æ·»åŠ æœªæ¥è¦é¢„æµ‹çš„3ä¸ªæœˆ
future = pd.DataFrame(index=date_list, columns= df_month.columns)
df_month2 = pd.concat([df_month2, future])
# ä»Žæœ€æ—©æ—¥æœŸ 1990-12-19 åˆ°çŽ°åœ¨
df_month2['forecast'] = best_model.predict(start=0, end=len(df_month2))
# ç¬¬ä¸€ä¸ªå…ƒç´ ä¸æ­£ç¡®ï¼Œè®¾ç½®ä¸ºNaN
df_month2['forecast'][0] = np.NaN
print(df_month2)
```

```python
plt.figure(figsize=(30,7))
df_month2.Price.plot(label="current")
df_month2.forecast.plot(color='r', ls='--', label='prediction')
plt.legend()
plt.title('shanghai monthly')
plt.xlabel('time')
plt.ylabel('points')
plt.show()
```

### LSTM 

```python
import numpy as np
import pandas as pd
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
from pandas import DataFrame
from pandas import concat
from itertools import chain
from sklearn.metrics import mean_squared_error
import matplotlib.pyplot as plt
```

```python
def get_train_set(data_set, timesteps_in, timesteps_out=1):
    train_data_set = np.array(data_set)
    reframed_train_data_set = np.array(series_to_supervised(train_data_set, timesteps_in, timesteps_out).values)
    print(reframed_train_data_set)
    print(reframed_train_data_set.shape)
    train_x, train_y = reframed_train_data_set[:, :-timesteps_out], reframed_train_data_set[:, -timesteps_out:]
    # å°†æ•°æ®é›†é‡æž„ä¸ºç¬¦åˆLSTMè¦æ±‚çš„æ•°æ®æ ¼å¼,å³ [æ ·æœ¬æ•°ï¼Œæ—¶é—´æ­¥ï¼Œç‰¹å¾]
    train_x = train_x.reshape((train_x.shape[0], timesteps_in, 1))
    return train_x, train_y

"""
å°†æ—¶é—´åºåˆ—æ•°æ®è½¬æ¢ä¸ºé€‚ç”¨äºŽç›‘ç£å­¦ä¹ çš„æ•°æ®
ç»™å®šè¾“å…¥ã€è¾“å‡ºåºåˆ—çš„é•¿åº¦
data: è§‚å¯Ÿåºåˆ—
n_in: è§‚æµ‹æ•°æ®input(X)çš„æ­¥é•¿ï¼ŒèŒƒå›´[1, len(data)], é»˜è®¤ä¸º1
n_out: è§‚æµ‹æ•°æ®output(y)çš„æ­¥é•¿ï¼Œ èŒƒå›´ä¸º[0, len(data)-1], é»˜è®¤ä¸º1
dropnan: æ˜¯å¦åˆ é™¤NaNè¡Œ
è¿”å›žå€¼ï¼šé€‚ç”¨äºŽç›‘ç£å­¦ä¹ çš„ DataFrame
"""
def series_to_supervised(data, n_in=1, n_out=1, dropnan=True):
    n_vars = 1 if type(data) is list else data.shape[1]
    df = DataFrame(data)
    cols, names = list(), list()
    # input sequence (t-n, ... t-1)
    for i in range(n_in, 0, -1):
        cols.append(df.shift(i))
        names += [('var%d(t-%d)' % (j + 1, i)) for j in range(n_vars)]
    # é¢„æµ‹åºåˆ— (t, t+1, ... t+n)
    for i in range(0, n_out):
        cols.append(df.shift(-i))
        if i == 0:
            names += [('var%d(t)' % (j + 1)) for j in range(n_vars)]
        else:
            names += [('var%d(t+%d)' % (j + 1, i)) for j in range(n_vars)]
    # æ‹¼æŽ¥åˆ°ä¸€èµ·
    agg = concat(cols, axis=1)
    agg.columns = names
    # åŽ»æŽ‰NaNè¡Œ
    if dropnan:
        agg.dropna(inplace=True)
    return agg


# ä½¿ç”¨LSTMè¿›è¡Œé¢„æµ‹
def lstm_model(source_data_set, train_x, label_y, input_epochs, input_batch_size, timesteps_out):
    model = Sequential()
    
    # ç¬¬ä¸€å±‚, éšè—å±‚ç¥žç»å…ƒèŠ‚ç‚¹ä¸ªæ•°ä¸º128, è¿”å›žæ•´ä¸ªåºåˆ—
    model.add(LSTM(128, return_sequences=True, activation='tanh', input_shape=(train_x.shape[1], train_x.shape[2])))
    # ç¬¬äºŒå±‚ï¼Œéšè—å±‚ç¥žç»å…ƒèŠ‚ç‚¹ä¸ªæ•°ä¸º128, åªè¿”å›žåºåˆ—æœ€åŽä¸€ä¸ªè¾“å‡º
    model.add(LSTM(128, return_sequences=False))
    model.add(Dropout(0.5))
    # ç¬¬ä¸‰å±‚ å› ä¸ºæ˜¯å›žå½’é—®é¢˜æ‰€ä»¥ä½¿ç”¨linear
    model.add(Dense(timesteps_out, activation='linear'))
    model.compile(loss='mean_squared_error', optimizer='adam')

    # LSTMè®­ç»ƒ input_epochsæ¬¡æ•°
    res = model.fit(train_x, label_y, epochs=input_epochs, batch_size=input_batch_size, verbose=2, shuffle=False)

    # æ¨¡åž‹é¢„æµ‹
    train_predict = model.predict(train_x)
    #test_data_list = list(chain(*test_data))
    train_predict_list = list(chain(*train_predict))

    plt.plot(res.history['loss'], label='train')
    plt.show()
    print(model.summary())
    plot_img(source_data_set, train_predict)

# å‘ˆçŽ°åŽŸå§‹æ•°æ®ï¼Œè®­ç»ƒç»“æžœï¼ŒéªŒè¯ç»“æžœï¼Œé¢„æµ‹ç»“æžœ
def plot_img(source_data_set, train_predict):
    plt.figure(figsize=(24, 8))
    # åŽŸå§‹æ•°æ®è“è‰²
    plt.plot(source_data_set[:, -1], c='b')
    # è®­ç»ƒæ•°æ®ç»¿è‰²
    plt.plot([x for x in train_predict], c='g')
    plt.legend()
    plt.show()

# è®¾ç½®è§‚æµ‹æ•°æ®input(X)çš„æ­¥é•¿ï¼ˆæ—¶é—´æ­¥ï¼‰ï¼Œepochsï¼Œbatch_size
timesteps_in = 3
timesteps_out = 3
epochs = 500
batch_size = 100
data = pd.read_csv('./currency_data/shanghai_index_1990_12_19_to_2020_03_12.csv')
data_set = data[['Price']].values.astype('float64')
# è½¬åŒ–ä¸ºå¯ä»¥ç”¨äºŽç›‘ç£å­¦ä¹ çš„æ•°æ®
train_x, label_y = get_train_set(data_set, timesteps_in=timesteps_in, timesteps_out=timesteps_out)

# ä½¿ç”¨LSTMè¿›è¡Œè®­ç»ƒã€é¢„æµ‹
lstm_model(data_set, train_x, label_y, epochs, batch_size, timesteps_out=timesteps_out)
```
